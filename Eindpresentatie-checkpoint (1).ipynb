```python
#Inladen Ic Opnames Data
ic_url_op = "https://data.rivm.nl/covid-19/COVID-19_ziekenhuis_ic_opnames_per_leeftijdsgroep.json"
req_ic_op = req.get(ic_url_op)
pd_ic_op = pd.json_normalize(req_ic_op.json())
display(pd_ic_op)
```

```python
### NaN weg
pd_ic_op = pd_ic_op.dropna()
```

```python
### Histogram aanmaken van het aantal IC opnames per leeftijdsgroep
fig = px.histogram(data_frame=pd_ic_op, x='Age_group', y='IC_admission', title='Histogram van het aantal IC opnames per leeftijdsgroep', 
                labels={'Age_group':'Leeftijdsgroep'})
fig.update_layout(yaxis_title='Som IC opnames')
fig.show()
```

```python
### Waardes met =-teken weghalen en onder de 1000
CSV = CSV[CSV['Vaccination_coverage_completed'].str.contains('=')==False]
CSV = CSV[CSV['Vaccination_coverage_completed'].astype(int) < 1000]
```

```python
### Alle NaN weg
GEO = GEO.dropna()
CSV = CSV.dropna()
CSV = CSV.dropna(subset = ['Region_name'])
CSV = CSV.dropna(subset = ['Vaccination_coverage_completed'])

### Gemiddelde uitrekenen Vaccination_coverage_completed en nieuwe kolom aanmaken genaamd Average
samen = GEO.copy()
samen['Average'] = None

def set_average(GEO_row):
    province = GEO_row['name']
    average = CSV[CSV['Province_name']==province]['Vaccination_coverage_completed'].astype(int).mean()
    GEO_row['Average']=average
    return GEO_row

samen = samen.apply(lambda row: set_average(row), axis=1)
    
average = CSV[CSV['Province_name']=='Friesland']['Vaccination_coverage_completed'].astype(int)

samen
```
